## Summations

Expressed as $\Sigma$, adds element together.

For example

$$
\sum_{5}^{i=1} 2i = (2)1 + (2)2 + (2)3 + (2)5 = 30
$$

In python can be expressed as:


```python
# i is placeholder for each consecutive index value
# range in python is end exclusive, meaning the latter value is not included
summation = sum(2*i for i in range(1,6))
print(summation)
```

Also common to see $n$ represent the number of items in collection:

$$
  \sum_{i=1}^{n} 10x_i
$$

```python
x = [1, 4, 6, 2]
# len is a built-in function that returns the length of an array
n = len(x)

summation = sum(10 * x[i] for i in range (0, n))
print(summation)
```

```python
# Summation in sympy
from sympy import *
i, n = symbols('i n')

# iterate each element i from 1 to n,
# then multiply and sum
summation = Sum(2*i, (i, 1, n))

# specify n as 5
up_to_5 = summation.subs(n, 5)
# do the summation
print(up_to_5.doit())
```

## Exponents

Exponents multiply a number by itself a specified number of times

$$
  2^{3} = 2 \times 2 \times 2 = 8
$$

Where $2$ is the *base* and $3$ is the *exponent*

Multiplication of exponent:

$$
x^2 x^3 = (x*x)*(x*x*x) = x^{2+3} = x^5
$$

Division of exponent:

$$
\frac{x^2}{x^5} = \frac{x*x}{x*x*x*x*x} = \frac{1}{x*x*x}
$$
$$
\frac{1}{x^3} = x^{-3}
$$



```python
# using sympy
from sympy import *

x = symbols('x')
expr = x**2 / x**5
print(expr)
```

Fractional exponents:

$$
4^{1/2} = \sqrt{4} = 2
$$

$$
8^{1/3} = \sqrt[3]{8} = 2 
$$

Power rule in exponents:

$$
(8^3)^2 = 8^{3 \times 2} = 8^6
$$

$$
8^{\frac{2}{3}} = (8^{\frac{1}{3}})^{2} = 2^2 = 4
$$

## Logarithms

A logarithm is a function that finds a power for a specific number and base.

Example of expression: "2 raised to *what power* gives me 8"?

$$
2^x = 8
$$

Can be expressed as 

$$
\log_{2} 8 = x
$$

where $2$ is the *base*

```python
from math import log

# 2 raised to what power gives me 8?
x = log(8, 2)

print(x)
```

Important note, when we don't give the *base* to the log function, Python will use Euler's number $e$ instead of base $10$.

## Euler's Number and Natural Logarithms

### Euler's Number

Why euler's number is used so much? A property of Euler’s number is its exponential function is a derivative to itself, which is convenient for exponential and logarithmic functions.

Watch [Numberphile - e(Euler's Number)](https://youtu.be/AuA2EAgAegE)

$e$ is the resulting value of the expression:

$$
(1 + \frac{1}{n})^{n}
$$

The larger $n$ value, we are getting closer to $2.71828$, which is the Euler's number.

### Natural Logarithms

When we use $e$ as the base for a logarithm, we call it *natural logarithm*. Rather than expressing natural logarithm as something like $\log_{e}10$, we would shorthand it as $\ln(10)$

```python
# Calculating the natural logaritm of 10 in Python
from math import log

# e raised to what power gives us 10?
x = log(10)
print(x)
```

## Limits

Given a function 

$$
f(x) = \frac{1}{x}
$$

As $x$ frover increases, $f(x)$ gets closer to 0 but never actually reaches 0; it just forver keeps getting closer, forever extends into infinity.

The way to express a value that is forver being approached but never reached is throught **limit**:

$$
\lim_{x \rightarrow	\infty} \frac{1}{x} = 0 
$$

which can be read "as $x$ approaches infinity, the function $1/x$ approaches 0 but never reaches 0".

```python
from sympy import *

x = symbols('x')
f = 1 / x
result = limit(f, x, oo)

print(result)
```

Since Euler's number $e$ is an irrational number (forever extending $n$ into infinity), can be expressed using:

$$
\lim_{n \rightarrow	\infty} (1 + \frac{1}{n})^n = e = 2.71828169413...
$$

```python
from sympy import *

n = symbols('n')
f = (1 + (1/n))**n
result = limit(f, n, oo)

# will returning E
print(result)

# will returning the number of E
print(result.evalf())
```

## Derivatives

A *derivative* tells the slope of a function, which useful to measure the rate of change at any point in a function.

>Why do we care about derivatives? They are often used in machine learning and other mathematical algorithms, especially with gradient descent. When the slope is 0, that means we are at the minimum or maximum of an output variable

We can measure "steepness" (*kecuraman*) at any point in the curve. 

To calculate the slope $m$ between two points using simple rise-over-run formula:

$$
m = \frac{y_2 - y_1}{x_2 - x_1}
$$


```python
# Manually calculating a derivative
def derivative_x(f, x, step_size):
  return (f(x + step_size) - f(x)) / step_size

def example_function(x):
  return x**2

slope_at_2 = derivative_x(example_function, 2, 0.0001)

print(slope_at_2)
```

When we encounter exponential function like $f(x) = x^2$, the derivative function will make the exponent a multipler then decrement the xponent by $1$, leaving us with the derivative:

$$
\frac{d}{dx}x^2 = 2x
$$

where $\frac{d}{dx}$ indicates a *derivative with respect to $x$*.

Let's say we wanted t ofind the slope at $x = 2$, 

```python
# Calculating derivatives using sympy
from sympy import *

x = symbols('x')

f = x**2

# Derivative of f with respect to x
df_dx = diff(f, x)
print(df_dx)
```
